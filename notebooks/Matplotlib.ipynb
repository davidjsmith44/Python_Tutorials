##### Importing Matplotlib
Python is very useful because you can load groups of functions created by other people into python and use their code. These groups of programs are called 'modules' in python (think of them as toolboxes that can do different things.) When we downloaded python it contained the most popular modules ('like Matplotlib')

Import statements cam be written as simply as this:
'import module_name'

Sometimes the module was written for a group of different programming languages and kept together. This is the case with matplotlib, so we need to import just the functions related to python. This is why we write the '.pyplot' in the import statement. (Just to tell python only to import the functions for python and not for R, Julia, or other programming languages).

Often, we want to give all of the functions in this module a suffix (also called an alias) to that we know we are using this specific module's functions. In our import statment, we are loading this module with the suffix 'plt'. If we want to use the plot function in matplotlib, we would write 'plt.plot()' to tell python we want it to use the plot function that was loaded from the matplotlib module.

##### Plots are objects
Plots are objects in python. This has two implications:
1. plots have access to many methods that can be used for all plots
    - set_title()
    - set_y_label()
    - set_x_label()
    - set_grid()
    -etc.
2. Plots are names like variables. This makes it very easy to change a plot's properties. For example, you name your plot then you can change it's colors, title or add icons or text to the plot.


##### Our first strange magic function
Python will usually create a plot, but it will not show it to you until you type 'name_of_plot.show()'. The '%matplotlib inline' statement tells python to show the chart anytime you create one. 'inline' means in the output space of the cell. So python will always show you the plot if you have this 2nd line below the import statement

###### Note: The two lines below will be at the top of any notebook you use in the future. It is always part of the import statements on my notebooks and you will type this out so many times, it will become second nature. (Like putting your name on a paper for school.)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Your first chart
We will create an list of numbers that are our x variables and a list of variables that are the y variables. 
##### Note: We need to make sure the length of x is equal to the length of y or python will not know what to do with the numbers

```python
x = [0,1,2,3,4,5,6]
y = [4.5, 2.3, 6.0, 1.4, 5, 2.7, 3.9]
```

```python
print('x is this long: ',len(x))
print('y is this long: ',len(y))
```

##### the plt.plot function needs two inputs:
 - x (list or array or other variable of numbers)
 - y (list or array or other variable of numbers)
 

```python
plt.plot(x,y)
```

### Now adding multiple lines to a plot
For this we are going to import numpy, a module (group of programs) that means numerical python. It has a lot of simple math functions that we can use. We import it as np (always use this) so when we want to use a numpy function, we call it with np.function_name (where function name is the actual name, like cos or sin)


```python
import numpy as np
```

#### Setting up a numpy array (just a list of numbers that are ordered, like a x-domain)
We are going to use np.arange. This function takes three inputs (start_value, end_value, step_size)

The default start_value is zero,
There is no default ending value, so if you call np.arange(8) it will give you a vector that goes from zero to 7
The default step size is 1


```python
x=np.arange(8)
x
```

```python
x = np.arange(-4,4)
x
```

```python
x = np.arange(-4,4,0.25)
x
```

#### Now lets get different functions of our x array to plot

```python
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
```

```python
plt.plot(x, sin_x)
plt.plot(x, cos_x)
plt.plot(x, tan_x)
```

#### OK, but lets dress this up a little by giving it a legend and labeling everything

We position the location of the legend, but specifying a loc variable

 - ===============   =============
 - Location String   Location Code
 -       ===============   =============
 -       'best'            0
 -       'upper right'     1
 -       'upper left'      2
 -       'lower left'      3

 -       'right'           5
 -       'center left'     6
 -       'center right'    7
 -       'lower center'    8

```python
plt.plot(x, sin_x, label='sin')
plt.plot(x, cos_x, label='cos')
plt.plot(x, tan_x, label='tan')
plt.legend(loc='upper center')
plt.title('My functions')
plt.xlabel('This is X')
plt.ylabel('This is y')
```

#### Lets make the X-axis bolder and maybe give it a grid and background color

```python
plt.plot(x, sin_x, label='sin')
plt.plot(x, cos_x, label='cos')
plt.plot(x, tan_x, label='tan')
plt.legend(loc='upper center')
plt.title('My functions')
plt.xlabel('This is X')
plt.ylabel('This is y')
plt.grid()
```

### StyleSheets THESE ARE VERY COOL
Matplotlib has standard stylesheets that contain styles for the charts (determine colors, background colors, grids, etc). You can see them by going to the web site below

[Go to the Matplotlib Style Sheets](https://matplotlib.org/3.2.1/gallery/style_sheets/style_sheets_reference.html)

All you have to do is place the line below at the start of the plot (or anywhere in the cell)

plt.style.use('name_of_style')

Lets Try a couple of these.

```python
plt.style.use('Solarize_Light2')
plt.plot(x, sin_x, label='sin')
plt.plot(x, cos_x, label='cos')
plt.plot(x, tan_x, label='tan')
plt.legend(loc='upper center')
plt.title('My functions')
plt.xlabel('This is X')
plt.ylabel('This is y')
```

```python
plt.style.use('ggplot')
plt.plot(x, sin_x, label='sin')
plt.plot(x, cos_x, label='cos')
plt.plot(x, tan_x, label='tan')
plt.legend(loc='upper center')
plt.title('My functions')
plt.xlabel('This is X')
plt.ylabel('This is y')
```

#### To make playing with the styles easlier, lets turn the plotting code into a function!

We will create a function that takes as an input three things
 1. The X values
 2. The Y values
 3. The stylesheet name
 
The function will just create the plot. We do not need it to return anything so we will put return None

```python
def plot_function(x, y, style_sheet):
    plt.plot(x, y)
    plt.style.use(style_sheet)
    plt.title('My functions')
    plt.xlabel('This is X')
    plt.ylabel('This is y')
    return None
```

```python
this_stylesheet = 'dark_background'
this_plot = plot_function(x, sin_x, this_stylesheet )
```

```python
this_stylesheet = 'classic'
this_plot = plot_function(x, sin_x, this_stylesheet)
```

```python
this_stylesheet = 'seaborn-dark'
this_plot = plot_function(x, tan_x, this_stylesheet)
```
