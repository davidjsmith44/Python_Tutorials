##### Importing Matplotlib
Python is very useful because you can load groups of functions created by other people into python and use their code. These groups of programs are called 'modules' in python (think of them as toolboxes that can do different things.) When we downloaded python it contained the most popular modules ('like Matplotlib')

Import statements cam be written as simply as this:
'import module_name'

Sometimes the module was written for a group of different programming languages and kept together. This is the case with matplotlib, so we need to import just the functions related to python. This is why we write the '.pyplot' in the import statement. (Just to tell python only to import the functions for python and not for R, Julia, or other programming languages).

Often, we want to give all of the functions in this module a suffix (also called an alias) to that we know we are using this specific module's functions. In our import statment, we are loading this module with the suffix 'plt'. If we want to use the plot function in matplotlib, we would write 'plt.plot()' to tell python we want it to use the plot function that was loaded from the matplotlib module.

##### Plots are objects
Plots are objects in python. This has two implications:
1. plots have access to many methods that can be used for all plots
    - set_title()
    - set_y_label()
    - set_x_label()
    - set_grid()
    -etc.
2. Plots are names like variables. This makes it very easy to change a plot's properties. For example, you name your plot then you can change it's colors, title or add icons or text to the plot.


##### Our first strange magic function
Python will usually create a plot, but it will not show it to you until you type 'name_of_plot.show()'. The '%matplotlib inline' statement tells python to show the chart anytime you create one. 'inline' means in the output space of the cell. So python will always show you the plot if you have this 2nd line below the import statement

###### Note: The two lines below will be at the top of any notebook you use in the future. It is always part of the import statements on my notebooks and you will type this out so many times, it will become second nature. (Like putting your name on a paper for school.)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Your first chart
We will create an list of numbers that are our x variables and a list of variables that are the y variables. 
##### Note: We need to make sure the length of x is equal to the length of y or python will not know what to do with the numbers

```python
x = [0,1,2,3,4,5,6]
y = [4.5, 2.3, 6.0, 1.4, 5, 2.7, 3.9]
```

```python
print('x is this long: ',len(x))
print('y is this long: ',len(y))
```

##### the plt.plot function needs two inputs:
 - x (list or array or other variable of numbers)
 - y (list or array or other variable of numbers)
 

```python
plt.plot(x,y)
```

##### OK we have a chart but it is not very interesting.
Can we use plot if our variables are not labeled x and y?

```python
not_labeled_x = x
not_labeled_y = y
plt.plot(not_labeled_x, not_labeled_y)
```

#### What if we switch the x and y by mistake?

```python
plt.plot(y,x)
```
